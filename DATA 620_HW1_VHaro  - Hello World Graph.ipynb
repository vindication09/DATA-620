We start by downloading the required packages to make our simple graph and creating an empty shell to be filled with our mock data. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
```

We need to define the nodes that will be used to make the graph. Once our nodes have bee defined, we then add them to our empty graph using the "add_nodes_from " command 

```python
nodes = ['Bart', 'Lisa', 'Maggie', 'Homer', 'Marge', 'Abe', 'Milhouse', 'Nelson', 'Ned']


G.add_nodes_from(nodes)
```

We need to define the edges. As of now, we have our nodes defined. Our task is to connect the nodes with lines known as edges.Mechanically, we define connections similar to how we define a pairwise (x,y) data point. We then append our edges to our graph system using the "add_from_edges" command. 

```python
edges = [('Bart', 'Lisa'),
         ('Bart', 'Maggie'),
         ('Bart', 'Homer'),
         ('Bart', 'Abe'),
         ('Lisa', 'Homer'),
         ('Lisa', 'Selma'),
         ('Lisa', 'Marge'),
         ('Maggie', 'Abe'),
         ('Maggie', 'Homer'),
         ('Marge', 'Homer'),
         ('Marge', 'Selma'),
         ('Homer', 'Abe'),
         ('Homer', 'Selma'),
         ('Abe','Selma'),
         ('Milhouse', 'Abe'),
         ('Milhouse', 'Selma'),
         ('Milhouse', 'Nelson'),
         ('Nelson','Ned' )
        ]


G.add_edges_from(edges)
```

Using the abilities within the networkx package, we can view the degree of each node. By definition, the degree is simply the number of edges connected to a node. As you will see in the output down below, Homer has the highest degree, where degree=6.

```python
nx.degree(G)
```

Lets construct our simple graph and see if it produces the desired Krackhardt kite

```python
nx.draw(G, with_labels = True)
plt.show()
```
